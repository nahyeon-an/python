```python
import os
import docx2txt
import pandas as pd

# customize
from pre_inc_doc import DocxDocument
from pre_inc_pdf import PdfDocument
```

```python
path = "/Users/nahyeonan/Downloads/incruit"
```

```python
# f = open('log.txt', 'w')

task_list = []
corp_list = []
question_list = []
```

```python
for file in os.listdir(path):
    ext = file[-3:]
    if ext == 'hwp':
        print('hwp')
        filename = path+"/"+file[:-4]+".txt"
        output = '--output ' + '"' + filename + '"'
        exefile = 'hwp5txt'
        result = '"' + path+'/'+file + '"'
        os.system(exefile + " " + output + " " + result)
```

```python
for file in os.listdir(path):
    ext = file[-3:]
    if ext == 'pdf':
        pdObj = PdfDocument(path+"/"+file)
        pdObj.extractQAList()
        pdObj.extractCorpName()
```

```python
for file in os.listdir(path):
    ext = file[-3:]
    if ext == 'ocx':
        text = docx2txt.process(path+"/"+file)
        lines = []
        for line in text.splitlines():
            if line == '\n' or line == '':
                continue
            lines.append(line)
        dd = DocxDocument(lines)
        corpName = dd.extractCorpName()
        task = dd.extractTask()
        q, a = dd.extractQAList()
        
        task_list.append(task)
        corp_list.append(corpName)
        question_list.append(q)
```
