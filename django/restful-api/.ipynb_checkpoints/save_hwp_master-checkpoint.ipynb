```python
from pre_inc_doc import DocxDocument

documents = []
for file in os.listdir("hwp"):
    f = open("hwp/"+file, 'rt')
    
    text_list = []
    for line in f.readlines():
        line.strip()
        if line == '\n' or line =='' or line == '<그림>\n' or line == ' ':
            continue
        text_list.append(line)
    
    f.close()
    
    temp = []
    if text_list[0][-6:] == '금합니다.\n':
        documents.append(text_list[1:])
        temp = text_list[1:]
    else:
        documents.append(text_list)
        temp = text_list
        
    dd = DocxDocument(temp)
    if len(temp) < 2:
        print(file)
        print(temp)
        continue
    corp = dd.extractCorpName()
    task = dd.extractTask()
    q, a = dd.extractQAList()
    
    task_list.append(task)
    corp_list.append(corp)
    question_list.append(q)
    
```
