# urllib : URL 처리 모듈

URL 작업을 위한 모듈을 모은 패키지

- urllib.request : url을 열고 읽기 위한 모듈
- urllib.error : request에 의해 발생하는 예외 처리
- urllib.parse : url 구문 분석
- urllib.robotparser : robots.txt 파일 구문 분석

## urllib.request


```python
from urllib import request
```

```python
req = request.urlopen("http://www.naver.com")
```

```python
type(req)
```

```python
req.close()
```

```python
with request.urlopen("http://www.naver.com/") as response:
    html = response.read()
```

```python
html
```

```python
type(html)
```

```python
data = html.decode('utf-8')
```

```python
data
```
